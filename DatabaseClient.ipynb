Import mysql.connector

```python
import mysql.connector as connector
```

Connect to the database

```python
db = {
  'host' : "127.0.0.1",
  'port' : 3306,
  'user' : "admin",
  'password' : "admin",
  'database' : "LittleLemonDB"
}
connection = connector.connect(**db)
```

List table names from the LittleLemonDB

```python
show_tables_query = "SHOW tables" 
cursor = connection.cursor()
cursor.execute(show_tables_query)
header = f"List of tables in the database {db['database']}:"
print(header)
print(len(header) * "-")
for record in cursor.fetchall():
    print(record[0])
```

The query:  
full name and contact details and total cost
for every customer that has placed an order greater than $60.

```python
customers_query = """
    select CONCAT (c.FirstName, ' ', c.LastName) AS FullName, c.PhoneNumber, c.Email, o.TotalCost
    FROM Customers c
    JOIN Orders o on o.CustomerID = c.CustomerID
    WHERE o.TotalCost > 60
"""
cursor.execute(customers_query)
```
